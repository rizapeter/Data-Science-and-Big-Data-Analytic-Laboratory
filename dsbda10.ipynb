```python
#importing required libraries
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
dataset = sns.load_dataset('iris')
```

# List down the features and their types (e.g., numeric, nominal) available in the dataset. Create a histogram for each feature in the dataset to illustrate the feature distributions.


***Inferecence:***

In the above dataset, we can observe that there are 5 features:
1. sepal_length of numeric type
2. sepal_width of numeric type
3. petal_length of numeric type
4. petal_width of numeric type
5. species of nominal type.

```python
sns.set_style("dark")
```

```python
fig , axes= plt.subplots(2,2, figsize=(16,9))
sns.histplot(dataset['sepal_length'], ax=axes[0,0])
sns.histplot(dataset['sepal_width'], ax=axes[0,1])
sns.histplot(dataset['petal_length'], ax=axes[1,0])
sns.histplot(dataset['petal_width'], ax=axes[1,1])
```

**Inference**

From the histograms plotted for the four features ('sepal_length', 'sepal_width', 'petal_length', 'petal_width') of the dataset:

1. Sepal Length ('sepal_length'):
The distribution of sepal lengths appears to be somewhat assymmetrically distributed, with the majority of observations clustering around the range of 5.0 to 6.5 cm. There are relatively fewer observations with sepal lengths below 4 cm or above 7-8 cm.
2. Sepal Width ('sepal_width'):
The distribution of sepal widths shows a slight skewness towards higher values, indicating that sepal widths tend to be more spread out towards the higher end of the scale.
The majority of observations fall within the range of approximately 2.0 to 4.5 cm.
3. Petal Length ('petal_length'):
The histogram for petal lengths reveals a bimodal distribution, suggesting the presence of two distinct groups or clusters within the data.
One cluster corresponds to shorter petal lengths, ranging from approximately 1 to 2.5 cm, while the other cluster represents longer petal lengths, ranging from around 3 to 6.5 cm.
4. Petal Width ('petal_width'):
Similar to petal length, the distribution of petal widths also exhibits a bimodal pattern, with two distinct peaks indicating the presence of two different groups.
One group corresponds to narrower petal widths, ranging from approximately 0.1 to 0.5 cm, while the other group represents wider petal widths, ranging from about 1.0 to 1.8 cm.

## **Create a boxplot for each feature in the dataset. Compare distributions and identify outliers**

```python
fig , axes= plt.subplots(2,2, figsize=(16,9))
sns.boxplot(x='species' , y='sepal_length', data=dataset, ax=axes[0,0])
sns.boxplot(x='species' , y='sepal_width', data=dataset, ax=axes[0,1])
sns.boxplot(x='species' , y='petal_length', data=dataset,  ax=axes[1,0])
sns.boxplot(x='species' , y='petal_width',  data=dataset, ax=axes[1,1])
```
