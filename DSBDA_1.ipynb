* sex: Represents the gender of the passenger (male or female).
* embarked: Represents the port of embarkation (C, Q, or S).

* embark_town: Represents the town of embarkation.

From the above summary of the data types of variables we can apply type conversions for:
Convert sex, embarked, who, embark_town, and alive from object to categorical.

```python
# Convert object variables to categorical
df[['sex', 'who', 'class']] = df[['sex', 'who', 'class']].astype('category')
```

#Step 6: Turn categorical variables into quantitative variables in Python.

Converting categorical variables into quantitative variables involves transforming categories into numerical representations that can be used in mathematical models or analyses. This conversion is necessary because many machine learning algorithms and statistical techniques require numerical input.

```python
# Perform one-hot encoding on categorical variables
encoded = pd.get_dummies(df, columns=['sex', 'who', 'class'])

# Display the first few rows of the encoded DataFrame
print(encoded.head())
```

Explanation:

pd.get_dummies(): This function is used to perform one-hot encoding on categorical variables. It creates binary columns for each category in the specified columns.

columns=['sex', 'embarked', 'who', 'embark_town', 'alive', 'class', 'deck']: This parameter specifies the columns to encode. Replace these column names with the categorical variables in your DataFrame.

df_encoded: This DataFrame contains the original data with categorical variables replaced by one-hot encoded columns.

By applying one-hot encoding, you transform categorical variables into a format that can be used as input in machine learning models or other numerical analyses. Each category is represented by a binary column, making it suitable for quantitative analysis.

```python
from sklearn.preprocessing import LabelEncoder

# Initialize LabelEncoder
label_encoder = LabelEncoder()
# Apply label encoding to each categorical variable using .loc
for column in ['sex', 'who', 'class']:
    df.loc[:, column] = label_encoder.fit_transform(df[column])

# Display the first few rows of the DataFrame with encoded categorical variables
print(df.head(10))
```

Three methods to convert categorical into numerical


1.   One-hot encoding
2.   Label encoding
