```python
#importinng required libraries
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#Loading the dataset from sns
data = sns.load_dataset('titanic')
```

```python
#checking the number of rows and columns
data.shape
```

```python
data.head()
```

```python
#to check null values
data.isnull().sum()
```

```python
#Handling missing values for age
mean_age = data['age'].mean()
data['age'].fillna(mean_age, inplace=True)
```

```python
#dist plot
sns.distplot(data['age'], bins = 10,kde=True)
```

```python
#joint plot
sns.jointplot(x = data['age'], y = data['fare'], kind ='hex')
```

```python
#rug plot
sns.rugplot(data['fare'])
```

```python
#  Plotting a histogram to visualize the distribution of ticket prices
plt.figure(figsize=(10, 6))
sns.histplot(data= data, x='fare', kde=True, bins=30,color='skyblue')
plt.title('Distribution of Ticket Prices')
plt.xlabel('Fare')
plt.ylabel('Frequency')
plt.show()
```

```python
#to check the distribution of fare for each passenger
sns.boxplot(x='class', y='fare', data=data)
plt.title('Distribution of Fare by Passenger Class')
plt.xlabel('Passenger Class')
plt.ylabel('Fare')
plt.show()
```
